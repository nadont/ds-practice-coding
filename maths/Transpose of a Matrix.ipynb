## Transpose of a Matrix
Write a Python function that computes the transpose of a given matrix.

```python
a = [[1,2,3],[4,5,6]]
```

```python
import numpy as np

A = np.array(a)
A.T
```

```python
result = [list(t) for t in zip(*a)]
result
```

```python
tp = [list(i) for i in zip(a[0], a[1])]
tp
```
