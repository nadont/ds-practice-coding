## Calculate 2x2 Matrix Inverse


Write a Python function that calculates the inverse of a 2x2 matrix. Return 'None' if the matrix is not invertible.

```python
matrix = [[4, 7], [2, 6]]
```

```python
import numpy as np
inverse = np.linalg.inv(matrix)
inverse
```

```python
a = matrix[0][0]
b = matrix[0][1]
c = matrix[1][0]
d = matrix[1][1]

det = (a*d - b*c)

adjoint = [[d, -b], [-c, a]]
adjoint
```

```python
inverse = [[(1/(det)*element) for element in row] for row in adjoint]
inverse
```

```python
def inverse_2x2(matrix: list[list[float]]) -> list[list[float]]:
	a = matrix[0][0]
	b = matrix[0][1]
	c = matrix[1][0]
	d = matrix[1][1]
	det = (a*d - b*c)

	adjoint = [[d, -b], [-c, a]]
	inverse = [[(1/(det)*element) for element in row] for row in adjoint]
	return inverse
```
