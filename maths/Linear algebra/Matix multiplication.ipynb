```python
## GPT
def matrixmul(a: list[list[int | float]],
              b: list[list[int | float]]) -> list[list[int | float]]:
    # Validate: A's columns must equal B's rows
    if len(a) == 0 or len(b) == 0 or len(a[0]) != len(b):
        return -1

    rows_a = len(a)
    cols_a = len(a[0])
    cols_b = len(b[0])

    # Make sure b is rectangular
    if any(len(row) != cols_b for row in b):
        return -1

    # Output matrix initialised with zeros
    c = [[0 for _ in range(cols_b)] for _ in range(rows_a)]

    # Matrix multiplication
    for i in range(rows_a):
        for j in range(cols_b):
            # Dot product of row i in A with column j in B
            total = 0
            for k in range(cols_a):
                total += a[i][k] * b[k][j]
            c[i][j] = total

    return c
```
